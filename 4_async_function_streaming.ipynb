```python

system_prompt = """
You are an expert at agentic AI. And now you help people by giving them guidelines based on their daily routine about how to learn agentic AI 
and instruct them for the same. You also prevent them from distracting and you query them about the same.
"""

user_query = "Tell me what to do everyday step by step to achieve this dream. I am currently working a full time job (9-6) it is remote. So sometimes I am free sometimes not"

message = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_query}
]

async def fetch_async_response():
    start_time = time.time()
    
    response = await asyncclient.chat.completions.create(
        model="gpt-4.1-mini",  # Or "gpt-4o" if async available
        messages=message
    )
    
    end_time = time.time()
    
    print("Async Response:\n", response.choices[0].message.content.strip())
    print("Time taken:", end_time - start_time)

# Run the async function
asyncio.run(fetch_async_response())
```

# 2. Streaming


```python
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()
```

```python
client = OpenAI()
```

```python
messages = [
    {
        "role":"user",
        "content":"Tell me a joke five times in a row"
    }
]
```

```python
response = client.responses.create(
    model = "gpt-4o",
    input = messages,
    stream = True
)

text = ''


for event in response:
    if event.type == 'response.output_text.delta':
        text += event.delta
        print(text)
```
